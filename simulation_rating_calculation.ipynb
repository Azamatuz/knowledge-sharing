```python
def wtc_generator(zero_one):
    wtc = []
    for i in range(5):
        wtc.append(random.choice(zero_one))
    return wtc

def rating_generator(one_ten):
    ratings = []
    for i in range(5):
        ratings.append(random.choice(one_ten))
    return ratings

def distribution(ratings, max_limit):
    dist = []
    for r in ratings:
        if r != min(ratings):
            dist.append(max_limit)
        else:
            dist.append(0)
    return dist

def wtc_change(dist, wtc):
    loc_wtc = []
    for key, val in enumerate(dist):
        if val == 0:
            loc_wtc.append(wtc[key] * 2)
        else:
            loc_wtc.append(wtc[key])
    return loc_wtc

def mat_rat_cal(ratings, loc_wtc):
    materials = []
    for key, val in enumerate(ratings):
        num_mat = int(loc_wtc[key] * 10)
        alfa = 0
        beta = 0
        gamma = 0
        
        for i in range(num_mat):
            alfa += random.choice(zero_one)
        for i in range(random.randint(1, 100)):
            beta += random.choice(zero_one)
        for i in range(random.randint(1, 100)):
            gamma += 0.1 
        materials.append(num_mat)
        ratings[key] = round(val + alfa + beta + gamma, 2)
    return materials, ratings
```
